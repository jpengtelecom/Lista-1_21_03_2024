```python
import numpy as np
import matplotlib.pyplot as plt

#gx = 0.5 1/(a-b)funcao conhecida

#c = 3 maxf(x)/g(x)
n=100000
xfunction=np.array([])

for i in range(n):
    x1=np.random.uniform(-1,1)
    x2=np.random.uniform(0,1)

    while (x2 >=(1.5*(x1**2))/1.5):
        x1=np.random.uniform(-1,1)
        x2=np.random.uniform(0,1)
    xfunction = np.append(xfunction, x1)

#print(xfunction)

X=np.arange(-1, 1, 0.1)
fx=1.5*(X**2)

plt.hist(xfunction,bins=100,density=True)
plt.grid(True)
plt.plot(X,fx)



```
