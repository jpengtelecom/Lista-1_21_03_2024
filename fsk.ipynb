```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special

t = np.linspace(0,50,2000)


signal1 = np.sin(2*np.pi*1*t)
signal2 = np.sin(2*np.pi*2.5*t)
data=(np.random.rand(10)>0.5)
data_sequence = np.repeat(data,200)

fsk = t*0


for x in range(len(t)):
    if data_sequence[x] == True:
        fsk[x] += signal2[x]
    else:
        fsk[x] += signal1[x]


plt.figure(1)
plt.plot(t, signal1)
plt.figure(2)
plt.plot(t, signal2)
plt.figure(3)
plt.plot(t,data_sequence)
plt.figure(4)
plt.plot(t,fsk)
plt.show()

```
