```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import special
from scipy import signal
from scipy import signal
import random

M = 4 #Ordem de modulação

f_s = 1000   # Sampling frequency
f = 5  # Frequency, in cycles per second, or Hertz
f_m = 2  # message frequency
a = 1 #amplitude
t = np.linspace(0,1000,f_s)


sinal_mensagem = np.sin(2*np.pi*f*t)

#portadora
x = np.arange(0, 1, 1/f_s)
portadora = a*signal.square(2*np.pi*f*x)

simbolos = np.floor(M*np.random.uniform(0,1,f_s))
pam = sinal_mensagem*simbolos

plt.figure(1)
plt.plot(t, sinal_mensagem)
plt.figure(2)
plt.plot(t, portadora)
plt.figure(3)
plt.plot(t, pam)
plt.show()

```
